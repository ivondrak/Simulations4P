## Å koda Superb Mass-Spring-Damper System

### Import of python modules

```python
from simulation_app import SimulationApp
```

### Disturbance force function definition. 
Force of 20N is going to disturb system from 0.2 to 1.5 seconds.

```python
def force_callback(t):
    if t >= 0.2 and t <= 1.5:
        return 20.0
    else:
        return 0.0
```

### Specification of the simulation parameters:

| Parameter | Description |
| --- | --- |
| title | title of output graph | 
| step | discretization of the simulation (time step - s) |  
| duration | duration of the simulation (s) | 
| mass | mass of the car (kg) | 
| damping | dampers coefficient (Ns/m) |
| stiffness | stiffnes of car spring (N/m) |

### Run the simulation

```python
title = 'Å koda Superb Mass-Spring-Damper System'
step = 0.01
duration = 4.0
mass = 1500.0
dumping = 4*3000.0
stiffness = 4*50000.0
app = SimulationApp(title, step,duration, mass, dumping, stiffness, force_callback)
app.run()
```
