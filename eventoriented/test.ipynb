```python
import sys
sys.path.append('..')

from simulation_app import SimulationApp, GeneratorFactory
from utilities.generator import ExponentialGenerator, UniformGenerator
# Exponential generator 
# Mean arrival rate = 3.0 mins
# Mean service rate = 1 min
class ExponentialFactory(GeneratorFactory):
    
    def createArrivals(self):
        return ExponentialGenerator(3, self.number_of_samples, self.seed)
    
    def createServices(self):
        return ExponentialGenerator(1, self.number_of_samples, self.seed)
    

# Uniform generator 
# Time between arrivals from 0 to 20 mins 
# Service time from 2 to 5 mins
class UniformFactory(GeneratorFactory):
    
    def createArrivals(self):
        return UniformGenerator(0, 20, self.number_of_samples, self.seed)
    
    def createServices(self):
        return UniformGenerator(2, 5, self.number_of_samples, self.seed)
    
title = 'Single queue single service simulation'
number_of_samples = 1000
app = SimulationApp(title, UniformFactory(number_of_samples))
app.run()
```
