## Testing statistics module

### Import modules

```python
import generator as gen
```

### Testing uniform generator

```python
g = gen.UniformGenerator(lower=0, upper=1, size=1000, seed=1959)
g.plot_distribution()
```

## Testing normal generator

```python
g = gen.NormalGenerator(mean=0, std=1, size=1000, seed=1959)
g.plot_distribution()
```

### Testing exponential generator

```python
g = gen.ExponentialGenerator(scale=0.8, size=1000, seed=1959)
g.plot_distribution()
```

### Testing Poisson generator

```python
g = gen.PoissonGenerator(lam=1, size=1000, seed=1959)
g.plot_distribution()
```

### Testing binomial generator 

```python
# Coin flip
g = gen.BinomialGenerator(n=1, p=0.5, size=1000, seed=1959)
g.plot_distribution()
```

### Testing logistic generator

```python
g = gen.LogisticGenerator(seed=1959)
g.plot_distribution()
```

### Testing multinomial generator

```python
# Dice roll
g = gen.MultinomialGenerator(n=1, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], size=1000, seed=1959)
g.plot_distribution()
```
